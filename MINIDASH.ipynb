```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt
FP=pd.read_csv("/content/drive/MyDrive/ames_data/03_eda.csv")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,9))
plt.suptitle("Ames Housing Distributions",fontsize=18,fontweight="bold")
#SalePrice
plt.subplot(2,2,1)
plt.hist(FP["SalePrice"],bins=30)
plt.title("SalePrice")
#Gr Liv Area
plt.subplot(2,2,2)
plt.hist(FP["Gr Liv Area"],bins=30)
plt.title("Gr Liv Area")
#Lot Area
plt.subplot(2,2,3)
plt.hist(FP["Lot Area"],bins=30)
plt.title("Lot Area")
#Overall Qual
plt.subplot(2,2,4)
plt.hist(FP["Overall Qual"],bins=10)
plt.title("Overall Qual")
plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```
